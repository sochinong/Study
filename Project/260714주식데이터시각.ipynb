```python
# FinanceDataReader 설치
!pip install finance-datareader

# 주가 데이터를 가져오기 위한 라이브러리
import FinanceDataReader as fdr

# 그래프를 그리기 위한 라이브러리
import matplotlib.pyplot as plt

# 오늘 날짜를 가져오기 위한 라이브러리
from datetime import datetime
```

```python
# 시작 날짜
start_date = "2021-01-01"

# 오늘 날짜를 문자열 형태(YYYY-MM-DD)로 저장
end_date = datetime.today().strftime("%Y-%m-%d")

# 삼성전자 주가 데이터 가져오기
df = fdr.DataReader("005930", start_date, end_date)

# 데이터 앞부분 5개 출력
df.head()
```

```python
# 그래프 크기 설정
plt.figure(figsize=(10, 5))

# 종가(Close) 그래프 그리기
plt.plot(df.index, df["Close"], label="Close")

# 그래프 제목
plt.title("Samsung Electronics Close Price")

# x축 라벨
plt.xlabel("Date")

# y축 라벨
plt.ylabel("Close")

# 범례 표시
plt.legend()

# 그래프 출력
plt.show()
```
