```python
import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import mplfinance as mpf
```

```python
import FinanceDataReader as fdr

# 삼성전자 주가 데이터 가져오기
df = fdr.DataReader(
    "005930",
    "2026-01-01",
    "2026-07-31"
)
```

```python
#삼전 데이터 가져오기
df = fdr.DataReader(
    "005930",
    "2026-01-01",
    "2026-07-31"
)

#시장 캔들차트 색상 설정
mc = mpf.make_marketcolors(
    up='red',
    down='blue',
    edge='inherit',
    wick='inherit',
    volume='inherit'
)

#스타일 생성
style = mpf.make_mpf_style(
    marketcolors=mc
)

#캔들차트 그리기
mpf.plot(
    df,
    type='candle',
    style=style,
    volume=True,
    title='Samsung Electronics (2026.01~2027.07)',
    ylabel='Price(KRW)',
    ylabel_lower='Volume'
)
```
